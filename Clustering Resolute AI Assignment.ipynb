```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("C:\\Users\\tekal\\OneDrive\\Rohan Data\\Intershalla\\Online Retail.xlsx - Online Retail.csv")
```

```python
data.isnull()
```

```python
data.isnull().sum().sum()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
plt.scatter(data['UnitPrice'],data['CustomerID'])
plt.xlim(-180,180)


plt.show
```

```python
x = data.iloc[:,5:7]
```

```python
x
```

```python
Kmeans = KMeans(2)
```

```python
Kmeans.fit(x)
```

```python
identified_cluster = Kmeans.fit_predict(x)
identified_cluster
```

```python
data_with_cluster = data.copy()
data_with_cluster['Cluster'] = identified_cluster
data_with_cluster
```

```python
plt.scatter(data_with_cluster['UnitPrice'],data_with_cluster['CustomerID'],c=data_with_cluster['Cluster'],cmap = 'rainbow')
plt.xlim(-180,180)
plt.show()
```
